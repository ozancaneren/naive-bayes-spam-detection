```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os             
mails_1 = os.listdir('enron1/all mails/')
mails_6 = os.listdir('enron6/all mails/')
x1 = len(mails_1)
x6 = len(mails_6)
```

```python
# Split data into two as "Train Set (70%)" and "Test Set (30%)" randomly
# Enron1

df_mails_1 = pd.DataFrame(np.zeros([x1,2]).astype(int), columns = ['filename', 'spam?'])

for i in np.arange(x1):
    df_mails_1.loc[i,'filename'] = mails_1[i]
    if df_mails_1.loc[i,'filename'].find('spam') >= 0:
        df_mails_1.loc[i,'spam?'] = 1

df_mails_1_train=df_mails_1.sample(frac=0.7)
df_mails_1_test=df_mails_1.drop(df_mails_1_train.index)
df_mails_1_train
```

```python
# Split data into two as "Train Set (70%)" and "Test Set (30%)"
# Enron6

df_mails_6 = pd.DataFrame(np.zeros([x6,2]).astype(int), columns = ['filename', 'spam?'])

for i in np.arange(x6):
    df_mails_6.loc[i,'filename'] = mails_6[i]
    if df_mails_6.loc[i,'filename'].find('spam') >= 0:
        df_mails_6.loc[i,'spam?'] = 1

df_mails_6_train=df_mails_6.sample(frac=0.7)
df_mails_6_test=df_mails_6.drop(df_mails_6_train.index)
df_mails_6_train
```

```python
# Find the unique words in train mails
# Enron1
import re

d1 = ''
for i in df_mails_1_train.index:
    a = df_mails_1_train.loc[i,'filename']
    with open(os.getcwd() + '/' + 'enron1' + '/' + 'all mails' + '/' + a, 'rt', errors='ignore' ) as fd:
        b = fd.read()
        c = re.sub(r'[^a-zA-Z]+', ' ' , b)
    d1 = d1+' '+c

d1 = d1.lower()
```

```python
len(d1.split())
```

```python
unique_words_1 = np.unique(d1.split())
e1 = len(unique_words_1)
e1
```

```python
# Find the unique words in train mails
# Enron6
import re

d6 = ''
for i in df_mails_6_train.index:
    a = df_mails_6_train.loc[i,'filename']
    with open(os.getcwd() + '/' + 'enron6' + '/' + 'all mails' + '/' + a, 'rt', errors='ignore' ) as fd:
        b = fd.read()
        c = re.sub(r'[^a-zA-Z]+', ' ' , b)
    d6 = d6+' '+c

d6 = d6.lower()
```

```python
len(d6.split())
```

```python
unique_words_6 = np.unique(d6.split())
e6 = len(unique_words_6)
e6
```

```python
# Create word count table
# Enron1
df_word_count_1 = pd.DataFrame(np.zeros([e1,2]).astype(int), columns = ['spam', 'ham'])
df_word_count_1.index = unique_words_1


for i in df_mails_1_train.index:
    a = df_mails_1_train.loc[i,'filename']
    f = 'ham'
    if df_mails_1_train.loc[i,'filename'].find('spam') >= 0:
        f = 'spam'
    with open(os.getcwd() + '/' + 'enron1' + '/' + 'all mails' + '/' + a, 'rt', errors='ignore' ) as fd:
        b = fd.read()
        c = re.sub(r'[^a-zA-Z]+', ' ' , b).lower()
        c = c.split()
        for j in c:
            df_word_count_1.loc[j,f] = df_word_count_1.loc[j,f] + 1

df_word_count_1
```

```python
# Add 0.1 to zero values in the "Word Count" table to prevent any probability value being zero
# Enron1
for i in df_word_count_1.index:
    for j in df_word_count_1.columns:
        if df_word_count_1.loc[i,j] == 0:
            df_word_count_1.loc[i,j] = df_word_count_1.loc[i,j] + 0.1

df_word_count_1
```

```python
# Create word count table
# Enron6
df_word_count_6 = pd.DataFrame(np.zeros([e6,2]).astype(int), columns = ['spam', 'ham'])
df_word_count_6.index = unique_words_6


for i in df_mails_6_train.index:
    a = df_mails_6_train.loc[i,'filename']
    f = 'ham'
    if df_mails_6_train.loc[i,'filename'].find('spam') >= 0:
        f = 'spam'
    with open(os.getcwd() + '/' + 'enron6' + '/' + 'all mails' + '/' + a, 'rt', errors='ignore' ) as fd:
        b = fd.read()
        c = re.sub(r'[^a-zA-Z]+', ' ' , b).lower()
        c = c.split()
        for j in c:
            df_word_count_6.loc[j,f] = df_word_count_6.loc[j,f] + 1

df_word_count_6
```

```python
# Add 0.1 to zero values in the "Word Count" table to prevent any probability value being zero
# Enron6
for i in df_word_count_6.index:
    for j in df_word_count_6.columns:
        if df_word_count_6.loc[i,j] == 0:
            df_word_count_6.loc[i,j] = df_word_count_6.loc[i,j] + 0.1

df_word_count_6
```

```python
# Create the "Word Probability" table from "Word Count" table
# Enron1
a = df_word_count_1['spam'].sum()
b = df_word_count_1['ham'].sum()
df_word_probability_1 = df_word_count_1.copy()
df_word_probability_1['spam'] = df_word_probability_1['spam'] / a
df_word_probability_1['ham'] = df_word_probability_1['ham'] / b
df_word_probability_1
```

```python
# Create the "Word Probability" table from "Word Count" table
# Enron6
a = df_word_count_6['spam'].sum()
b = df_word_count_6['ham'].sum()
df_word_probability_6 = df_word_count_6.copy()
df_word_probability_6['spam'] = df_word_probability_6['spam'] / a
df_word_probability_6['ham'] = df_word_probability_6['ham'] / b
df_word_probability_6
```

```python
# Calculate Spam/Ham probabilities
# Enron1
from mpmath import mpf

df_mails_1_test['P(Spam)'] = np.ones(len(df_mails_1_test))
df_mails_1_test['P(Ham)'] = np.ones(len(df_mails_1_test))

x = df_word_count_1['spam'].sum()
y = df_word_count_1['ham'].sum()
x = x/(x+y)
y + 1-x

for i in df_mails_1_test.index:
    a = df_mails_1_test.loc[i,'filename']

    with open(os.getcwd() + '/' + 'enron1' + '/' + 'all mails' + '/' + a, 'rt', errors='ignore' ) as fd:
        b = fd.read()
        c = re.sub(r'[^a-zA-Z]+', ' ' , b).lower()
        c = c.split()
        for j in c:
            if df_word_probability_1.index.isin([j]).sum() >=1:
                df_mails_1_test.loc[i,'P(Spam)'] = mpf(df_mails_1_test.loc[i,'P(Spam)']) * df_word_probability_1.loc[j,'spam']
                df_mails_1_test.loc[i,'P(Ham)']  = mpf(df_mails_1_test.loc[i,'P(Ham)']) * df_word_probability_1.loc[j,'ham']
        df_mails_1_test.loc[i,'P(Spam)'] = mpf(df_mails_1_test.loc[i,'P(Spam)']) * x
        df_mails_1_test.loc[i,'P(Ham)']  = mpf(df_mails_1_test.loc[i,'P(Ham)'])  * y

    df_mails_1_test.loc[i,'P(Spam)'] = mpf(df_mails_1_test.loc[i,'P(Spam)']) / mpf((mpf(df_mails_1_test.loc[i,'P(Spam)']) + mpf(df_mails_1_test.loc[i,'P(Ham)'])))
    df_mails_1_test.loc[i,'P(Ham)']  = 1 - mpf(df_mails_1_test.loc[i,'P(Spam)'])

df_mails_1_test
```

```python
# Calculate Spam/Ham probabilities
# Enron6
from mpmath import mpf

df_mails_6_test['P(Spam)'] = np.ones(len(df_mails_6_test))
df_mails_6_test['P(Ham)'] = np.ones(len(df_mails_6_test))

x = df_word_count_6['spam'].sum()
y = df_word_count_6['ham'].sum()
x = x/(x+y)
y + 1-x

for i in df_mails_6_test.index:
    a = df_mails_6_test.loc[i,'filename']

    with open(os.getcwd() + '/' + 'enron6' + '/' + 'all mails' + '/' + a, 'rt', errors='ignore' ) as fd:
        b = fd.read()
        c = re.sub(r'[^a-zA-Z]+', ' ' , b).lower()
        c = c.split()
        for j in c:
            if df_word_probability_6.index.isin([j]).sum() >=1:
                df_mails_6_test.loc[i,'P(Spam)'] = mpf(df_mails_6_test.loc[i,'P(Spam)']) * df_word_probability_6.loc[j,'spam']
                df_mails_6_test.loc[i,'P(Ham)']  = mpf(df_mails_6_test.loc[i,'P(Ham)']) * df_word_probability_6.loc[j,'ham']
        df_mails_6_test.loc[i,'P(Spam)'] = mpf(df_mails_6_test.loc[i,'P(Spam)']) * x
        df_mails_6_test.loc[i,'P(Ham)']  = mpf(df_mails_6_test.loc[i,'P(Ham)'])  * y

    df_mails_6_test.loc[i,'P(Spam)'] = mpf(df_mails_6_test.loc[i,'P(Spam)']) / mpf((mpf(df_mails_6_test.loc[i,'P(Spam)']) + mpf(df_mails_6_test.loc[i,'P(Ham)'])))
    df_mails_6_test.loc[i,'P(Ham)']  = 1 - mpf(df_mails_6_test.loc[i,'P(Spam)'])

df_mails_6_test
```

```python
# Make spam prediction
# Enron1
df_mails_1_test['spam?(predicted)'] = np.zeros(len(df_mails_1_test)).astype(int)
df_mails_1_test['PredictionTrue?'] = np.zeros(len(df_mails_1_test)).astype(int)

for i in df_mails_1_test.index:
    if df_mails_1_test.loc[i,'P(Spam)'] > df_mails_1_test.loc[i,'P(Ham)']:
        df_mails_1_test.loc[i,'spam?(predicted)'] = 1
    if df_mails_1_test.loc[i,'spam?'] == df_mails_1_test.loc[i,'spam?(predicted)']:
        df_mails_1_test.loc[i,'PredictionTrue?'] = 1
            
df_mails_1_test
```

```python
# Make spam prediction
# Enron6
df_mails_6_test['spam?(predicted)'] = np.zeros(len(df_mails_6_test)).astype(int)
df_mails_6_test['PredictionTrue?'] = np.zeros(len(df_mails_6_test)).astype(int)

for i in df_mails_6_test.index:
    if df_mails_6_test.loc[i,'P(Spam)'] > df_mails_6_test.loc[i,'P(Ham)']:
        df_mails_6_test.loc[i,'spam?(predicted)'] = 1
    if df_mails_6_test.loc[i,'spam?'] == df_mails_6_test.loc[i,'spam?(predicted)']:
        df_mails_6_test.loc[i,'PredictionTrue?'] = 1
            
df_mails_6_test
```
